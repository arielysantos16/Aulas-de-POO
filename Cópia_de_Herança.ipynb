# Pilares da OO: Herança

> **Herança**: Permite criar uma classe a partir de outra já existente, sendo a classe derivada (subclasse) herdeira de todos os atributos e métodos da classe principal (superclasse).

## O que já conhecemos: classes independentes

Imagine o sistema de uma empresa que registra nome, CPF e salário de todos os seus funcionários. Os funcionários da gerência são os únicos que tem acesso ao sistema, e portanto login e senha.

No início, vamos criar as duas classes **sem herança**, como se fossem independentes.

````python
## classe 1: Funcionário é a super classe
class Funcionario:
    def __init__(self, nome, cpf, salario):
        self.nome = nome
        self.cpf = cpf
        self.salario = salario

    def imprimir_dados(self):
        print(f"Funcionário: {self.nome} ({self.cpf})")
        print(f"Salário: R$ {self.salario:.2f}")

## classe 2: Gerente
class Gerente(Funcionario): # é uma subclasse de funcionario
    def __init__(self, nome, cpf, salario, login, senha):
        super().__init__(nome, cpf, salario)
        self.login = login
        self.senha = senha

    def imprimir_dados(self):
        super().imprimir_dados()
        print(f"Login: {self.login}, senha: {self.senha}")


    def validar_login(self, login, senha):
        return login == self.login and senha == self.senha

## Manipulando objeto do tipo Funcionário
f1 = Funcionario('Julia', '010.212.323-45', 5430.79)
f1.imprimir_dados()
print('-'*30)

## Manipulando objeto do tipo Gerente
g1 = Gerente('Aline', '002.323.434-56', 10430.79, 'line', 'line123')
g1.imprimir_dados()
````

## Herança: relação "é um"

As duas classes possuem:

- atributos `nome`, `CPF`, `salario`
- método `imprimir_dados()`

Faz sentido para o problema afirmar que todo **`Gerente` é um `Funcionário`**. Essa relação é refletida no código pela herança.

Nessa formulação, um gerente é uma versão mais especializada de funcionário, ou seja:

- `Funcionario` → Superclasse
- `Gerente` → Subclasse

Vamos reescrever as classes fazendo `Gerente` herdar de `Funcionario`.


````python
## Superclasse: Funcionário
class Funcionario:
    def __init__(self, nome, cpf, salario):
        self.nome = nome
        self.cpf = cpf
        self.salario = salario

    def imprimir_dados(self):
        print(f"Funcionário: {self.nome} ({self.cpf})")
        print(f"Salário: R$ {self.salario:.2f}")

## Subclasse: Gerente
class Gerente(Funcionario): # nome da superclasse entre parênteses na definição
    def __init__(self, nome, cpf, salario, login, senha):
        ### inicializando atributos da superclasse
        super().__init__(nome, cpf, salario)

        ### inicializando meus próprios atributos
        self.login = login
        self.senha = senha

    def imprimir_dados(self):
        super().imprimir_dados() # chamando método da superclasse
        print(f"Login: {self.login}, senha: {self.senha}")

    def validar_login(self, login, senha):
        return login == self.login and senha == self.senha


## Manipulando objeto do tipo Funcionário
f1 = Funcionario('Julia', '010.212.323-45', 5430.79)
f1.imprimir_dados()
print('-'*30)

## Manipulando objeto do tipo Gerente
g1 = Gerente('Aline', '002.323.434-56', 10430.79, 'line', 'line123')
g1.imprimir_dados()
print('Tentativa de login: ', g1.validar_login('aline',  '123456')) # retorna False
````

### O que mudou?

- A definição da classe
```python
class Subclasse(Superclasse)`
```

- Não precisamos redefinir na subclasse os atributos `nome`, `CPF` e `salario`, herdamos da superclasse.
   - Também não precisamos redefinir o método `imprimir_dados()` a menos que ele tenha um comportamento específico. **Experimente deletá-lo da subclasse**.

- Usamos o comando `super()` para acessar elementos da superclasse, inclusive o cosntrutor

```python
super().__init__(nome, cpf salario)
```

## Encapsulamento: níveis de acesso/visibilidade

Agora vamos falar sobre algo importante: **quem pode acessar os atributos no caso da herança?**

Vamos relembrar as convenções de nomenclatura do Python:

* `var_name`: **público**. Variáeis declaradas sem nenhum underline no início podem ser publicamente acessadas.

* `_um_underline`: **protegido**. Indicador fraco de "uso interno". Por convenção


* `__dois_underlines`: **privado**. Ao nomear um atributo ou método de classe,
provoca a alteração de nomes (dentro da classe Exemplo,
__x torna-se _Exemplo__x).

Guia de estilos do Python: https://peps.python.org/pep-0008/#descriptive-naming-styles

````python
class Funcionario:
    def __init__(self, nome, cpf, salario):
        self._nome = nome # protegido (sub pode acessar)
        self._cpf = cpf   # protegido
        self._salario = salario # protegido

    def imprimir_dados(self):
        print(f"Funcionário: {self._nome} ({self._cpf})")
        print(f"Salário: R$ {self._salario:.2f}")


class Gerente(Funcionario):
    def __init__(self, nome, cpf, salario, login, senha):
        super().__init__(nome, cpf, salario)

        self.__login = login  # privado
        self.__senha = senha  # privado

    def boas_vindas(self,):
        # subclasses podem acessar atributos protegidos da super
        print(f'Olá {self._nome}')

    def validar_login(self, login, senha):
        return login == self.login and senha == self.senha

f1 = Funcionario('Julia', '010.212.323-45', 5430.79)
f1.imprimir_dados()

g1 = Gerente('Aline', '002.323.434-56', 10430.79, 'line', 'line123')
g1.boas_vindas()
````

### Q1

Você deve criar um sistema simples de pet virtual através da seguinte modelagem:
- Superclasse `Pet` possui atributos protegidos `nome` e `humor, além dos métodos:
   -  `mostrar_status()`→ Imprime os dados da classe
   -  `brincar()` → Aumenta +1 no humor
- Subclasse `CaoRobo` possui atributo específico `carga_bateria` além do método:
   - `carregar_bateria(valor: int)` → aumenta a bateria +valor e aumenta +1 no humor
- Subclasse `Dragao` possui atributo específico `nivel_fogo` além do método:
   - `soltar_fogo()` → aumenta o nível do fogo +1 e diminui -1 no humor

<img src="https://drive.google.com/uc?id=1V1J9XEq4kFqyjK5yr2hcp8eo2JHrXVkd" width=650>


````python
# programe aqui
class Pet():
  def __init__(self, nome, humor):#está como opcional o valor do humor. Se o usuário adicionar algo o valor muda, se não ele sempre será 10
    self._nome = nome
    self._humor = humor

  def mostrar_status(self):
    print(f"O seu pet {self._nome} possui o humor {self._humor}", end=' ')#mantém o próximo texto na msm linha

  def brincar(self):
    self._humor += 1     #append é somente em listas


class CaoRobo(Pet):
  def __init__(self, nome, humor=5, carga_bateria=50): ##
    super().__init__(nome, humor)
    self.__carga_bateria = carga_bateria

  def carregar_bateria(self, valor):
    self.__carga_bateria += valor
    self._humor += 1

  def imprimir_dados(self):
    super().mostrar_status()
    print(f"e a bateria do seu pet está em {self.__carga_bateria}%.")


class Dragao(Pet):
  def __init__(self, nome, humor=5, nivel_fogo=0):
    super().__init__(nome, humor)
    self.__nivel_fogo = nivel_fogo

  def soltar_fogo(self):
    self.__nivel_fogo += 1
    self._humor -= 1

  def imprimir_dados(self):
    super().mostrar_status()
    print(f"e o nível de fogo do seu dragão está em {self.__nivel_fogo}.")
````

### Q3
Escreva uma classe **`Conta`** com atributo saldo (float), além dos seguintes métodos:
* `depositar(valor)`: Acrescentar o valor recebido como parâmetro ao atributo saldo
* `sacar(valor)`: Subtrair o valor recebido como parâmetro do atributo saldo **se houver saldo suficiente**.

Escreva uma subclasse de Conta chamada **`Poupanca`**. A classe Poupanca deve ter um atributo chamado `taxaRendimento` do tipo float que armazena percentual de rendimento ao mês da poupança. Crie um método `render()` que calcula o rendimento da poupança e atualiza o saldo.

Escreva uma subclasse de Conta chamada **`ContaEspecial`**, que possui um atributo adicional `limite` (float). Implemente o método `sacar(valor)` de forma que:
* Primeiro, o valor do saque seja descontado do saldo, mas o saldo nunca fica negativo (mínimo 0);
* Caso o valor do saque ultrapasse o saldo disponível, o restante deve ser descontado do limite, **desde que a soma do saldo com o limite seja suficiente** para cobrir o valor total do saque;


> * Use encapsulamento nas duas classes, ou seja, atributos devem ser privados para subclasses e protegidos para a superclasse
> * O construtor de cada classe deve receber todos os atributos de inicialização.


````python
class Conta:
  def __init__(self, saldo=0):
    self.saldo = saldo

  def depositar(self, deposito):
    self.saldo += deposito
    return self.saldo

  def sacar(self, saque):
    if (saque <= self.saldo):
      self.saldo -= saque
      print(f"Seu saque já foi efeatuado! \n Saldo atual é de R${self.saldo}.")
      return True
    elif (saque > self.saldo):
      print("Não é possível sacar um valor maior do que o seu saldo. :(")
      return False



class Poupanca(Conta):
  def __init__(self, saldo=0):
    super().__init__(saldo)

    self.__taxaRendimento = 0

  def render(self):
    self.__taxaRendimento += (self.__taxaRendimento * 0.03)
    return self.__taxaRendimento, self.saldo



class ContaEspecial(Conta):
  def __init__(self, saldo=0):
     super().__init__(saldo)
     self.__limite = 15000

  def movimentar(self, saque):
    retorno_da_super = super().sacar(saque)#o próprio sacar da superclasse faz esse cálculo

    if retorno_da_super == False:
      print("Porém...você tem uma conta ESPECIAL! \n Vamos ver o que podemos fazer por você. <3")
      if (self.saldo + self.__limite) >= saque :#pergunta se a soma deles é suficiente p/ cobrir o saque
        __restante = (saque - self.saldo)#pergunta quanto falta pra pagar
        if (__restante > 0):
          self.__limite = (self.__limite - __restante)
          self.saldo = 0
          print(f"Seu saque já foi efeatuado! \nVocê retirou do seu limite R${__restante}, seu saldo atual é de R${self.saldo} e seu limite atual é de R${self.__limite}.")

        else:
          print("É possível realizar o seu saque. \nSeu saque já foi efeatuado!")
          print(f"Saldo atual:{self.saldo}")

      else:
        print("Opa... você estorou o limite da sua conta.  :/ \n Seu saque não será possível de realizar.")



p1 = Poupanca()
p1.depositar(20)
p1.render()
p1.sacar(10)

ce1 = ContaEspecial(30)
ce1.depositar(50)
ce1.sacar(90)
ce1.movimentar(90)
````
